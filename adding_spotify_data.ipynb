```python
import pandas as pd
```

```python
data_final = pd.read_csv('audio_features.csv')
data_final.head(10)
```

```python
data_final[['artist_name', 'track_name']] = data_final['name'].str.extract(r'(.*)- (.*)')

```

```python
data_final['artist_name'] = data_final['artist_name'].str.strip()
data_final['track_name'] = data_final['track_name'].str.strip()
data_final[['artist_name', 'track_name']]
```

```python
data_enriquecida = pd.read_csv('Data_completa_enriquecida.csv')
data_enriquecida
```

```python
data_enriquecida.columns
```

```python
data_enriquecida['genre'].head()
```

```python
to_drop = [
    'Letra',
    'Idioma',
    'artist_uri',
    'cuadrante',
    'Lyrics',
    'cuadReal',
    'lexicones',
    'lyrics_state',
    'emociones',
    'emocionesLetra',
    'emocion_mas_comun',
    'dataCancion',
    'hasLetra',
    'spotify_id',
    'number_of_emotion_tags',
    'lastfm_url',
]

data_enriquecida = data_enriquecida.drop(to_drop, axis=1)
```

```python
from fuzzywuzzy import fuzz
def fuzzy_match(row):
    artist_similarity = fuzz.ratio(row['artist_name_x'], row['artist_name_y'])
    track_similarity = fuzz.ratio(row['track_name_x'], row['track_name_y'])
    return artist_similarity >= 90 and track_similarity >= 90
```

```python

df_merged = data_final.merge(data_enriquecida, on=['artist_name', 'track_name'], how='inner', left_on=None, right_on=None,
         left_index=False, right_index=False, sort=False,
         suffixes=('_x', '_y'), copy=True, indicator=True,
         validate=None)

df_merged
```

```python
def clean_column(value):
    # Check if the value is a string and contains brackets
    if isinstance(value, str) and value.startswith('[') and value.endswith(']'):
        # Extract the number between the brackets
        return value.strip('[]')
    return value  # Return the
```

```python
df_merged['tempo_x'] = df_merged['tempo_x'].apply(clean_column)
df_merged['tempo_x'] = pd.to_numeric(df_merged['tempo_x'])
df_merged
```

```python
df_merged.to_csv('final_data.csv')
```
