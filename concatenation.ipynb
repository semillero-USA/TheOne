## Using pandas for the different csv concatenation

```python
import pandas as pd
import numpy as np
```

- Reading the data

```python
csvs = []
csvs.append(pd.read_csv('franklin_audio_features.csv'))
csvs.append(pd.read_csv('mateo_audio_features.csv'))
csvs.append(pd.read_csv('santiago_audio_features.csv'))
csvs.append(pd.read_csv('sebastian_audio_features.csv'))
csvs.append(pd.read_csv('isabella_audio_features.csv'))
```

- Showing the data

```python
final_data = pd.concat([x for x in csvs])
final_data
```

```python
final_data.shape
```

```python
final_data = final_data.sort_values("name")
final_data
```

```python
final_data  = final_data.dropna()
final_data
```

```python
final_data = final_data.reset_index().drop("index", axis=1)
final_data
```

```python
final_data.to_csv("audio_features.csv")
```
